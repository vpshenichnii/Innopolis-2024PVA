```python
def get_outlier_boundaries(df, variable = 'item_price'):
    q1 = df[variable].quantile(0.25)
    q3 = df[variable].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    return [lower_bound, upper_bound]
```

```python
get_outlier_boundaries(df_all)
```

```python
df_all.groupby('item_category_name').apply(get_outlier_boundaries, include_groups=False)
```

```python
df_all.groupby('item_category_name')[['item_category_name', 'item_price']].apply(get_outlier_boundaries)
```

```python
def drop_outliers(df, variable = 'item_price'):
    q1 = df[variable].quantile(0.25)
    q3 = df[variable].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    return df[(df['item_price']>=lower_bound)&(df['item_price']<=upper_bound)]
```

```python
drop_outliers(df_all)
```

```python
(
    df_all
    .groupby('item_category_name')
    .apply(drop_outliers, include_groups=False)
    .reset_index(drop = False).drop('level_1', axis = 1)
)
```

### Метод TRANSFORM

```python
def get_lower_boundary(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    return lower_bound

def get_upper_boundary(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    return upper_bound
```

```python
# Работает как apply по сгруппированным данным, но возвращает результат в виде исходного DF 
# Похоже на оконные функции в SQL
df_all['lb'] = df_all.groupby('item_category_name')['item_price'].transform(get_lower_boundary)
df_all['ub'] = df_all.groupby('item_category_name')['item_price'].transform(get_upper_boundary)
```
