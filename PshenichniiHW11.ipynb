```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_penguins = pd.read_csv(r'https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Penguins.csv')
```

```python
df_penguins.head()
```

```python
sns.set(palette = 'deep', context = 'poster')
```

```python
plt.figure(figsize = (15, 7))
sns.barplot(data = df_penguins, x = 'species', y = 'body_mass_g', hue='sex')
plt.show() 
```

### Видим среднюю массу пингвинов по видам и полу. Видно что Gentoo самые крупные. У всех самцы крпнее самок.

### График отображает зависимость длины плавника от массы пингвина. Размер точки показывает размер клюва. Предстазуемо более тяжелые пингвины имеют больший клюв.

```python
df = df_penguins[df_penguins['sex'].isin(['MALE', 'FEMALE'])]
df = df.groupby(['species', 'sex'])['body_mass_g'].mean().unstack()/1000
sns.heatmap(df, cmap = 'YlGnBu', annot = True, fmt = ".2f")

#sns.heatmap(df_penguins, x = 'body_mass_g', y = 'island', cmap = 'YlGnBu', annot = True, fmt = ".2f")
```
