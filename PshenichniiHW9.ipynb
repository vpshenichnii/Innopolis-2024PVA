```python
import pandas as pd
import numpy as np
```

```python
df_penguins = pd.read_csv(r'https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Penguins.csv')
```

```python
def get_lower_boundary(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    return lower_bound

def get_upper_boundary(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    return upper_bound
```

```python
df_penguins['lb'] = df_penguins.groupby('sex')['body_mass_g'].transform(get_lower_boundary)
df_penguins['ub'] = df_penguins.groupby('sex')['body_mass_g'].transform(get_upper_boundary)
```
