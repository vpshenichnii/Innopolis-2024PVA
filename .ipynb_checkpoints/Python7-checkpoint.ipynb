### Заполнение пропусков для числовых переменных

```python
na_index = df_sales[df_sales['item_price'].isna()].index
```

```python
mean_price = df_sales['item_price'].mean()
df_sales['item_price_filled_mean'] = df_sales['item_price'].fillna(mean_price)
df_sales.loc[na_index, 'item_price_filled_mean']
```

```python
median_price = df_sales['item_price'].median()
df_sales['item_price_filled_median'] = df_sales['item_price'].fillna(median_price)
df_sales.loc[na_index, 'item_price_filled_median']
```

```python
df_sales['item_price_filled_arb'] = df_sales['item_price'].fillna(-999)
df_sales.loc[na_index, 'item_price_filled_arb']
```

```python
size_na = df_sales['item_price'].isna().sum()
random_prices = df_sales['item_price'].dropna().sample(size_na).values
df_sales.loc[df_sales['item_price'].isna(), 'item_price_filled_rand'] = random_prices 
```

```python
df_sales.loc[na_index, 'item_price_filled_rand']
```

### Заполнение пропусков для категориальных переменных
