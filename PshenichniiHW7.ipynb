### выбросы

```python
mu = df_union['body_mass_g'].mean()
sigma = df_union['body_mass_g'].std()
lower_bound = mu - 3*sigma
upper_bound = mu + 3*sigma
print(f"Выбросами будут считаться все массы пингвинов, которые находится за пределами интервала [{lower_bound}, {upper_bound}]")
```

```python
df_no_outliers = df_union[(df_union['body_mass_g']>=lower_bound)&(df_union['body_mass_g']<=upper_bound)]
```

```python
df_no_outliers
```

### нормализация

```python
df_no_outliers['norm_body_mass_g'] = (df_no_outliers['body_mass_g'] - df_no_outliers['body_mass_g'].min()) / (df_no_outliers['body_mass_g'].max() - df_no_outliers['body_mass_g'].min()) 
```
