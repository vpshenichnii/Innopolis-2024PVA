### GroupBy + Agg 

```python
(
    df_shops
    .groupby('year')['sales']
    .agg('sum')
)
```

```python
(
    df_shops
    .groupby('year')
    .agg({'sales': 'sum'})
)
```

```python
(
    df_shops
    .groupby('year')
    .agg({'sales': ['sum', 'mean', 'min', 'max', 'count']})
)
```

```python
(
    df_shops
    .groupby(['year', 'shop_name'])
    .agg({'sales': ['sum', 'mean', 'min', 'max', 'count']})
)
```

```python
(
    df_shops
    .groupby(['year', 'shop_name'])
    .agg({'sales': [('Сумма', 'sum'), ('Cреднее', 'mean'), 'min', 'max', 'count']})
)
```

```python
(
    df_shops
    .groupby(['year', 'shop_name'])
    .agg(
        {
            'sales': [('Сумма', 'sum'), ('Cреднее', 'mean')],
            'quantity' : [('Минимум', 'min'), ('Значения', 'unique')]
        }
    )
)
```
