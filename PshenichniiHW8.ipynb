```python
import pandas as pd
import numpy as np
```

```python
df_penguins = pd.read_csv(r'https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Penguins.csv')
```

```python
df_penguins
```

```python
clear_df_penguins = df_penguins.dropna().copy()
```

```python
clear_df_penguins
```

### Широкий т.к. несколько числовых значений 

```python
clear_df_penguins.stack().reset_index().melt()
```

```python
(
    clear_df_penguins
    .reset_index()
    .melt(
        id_vars = 'body_mass_g', 
        value_vars = ['island'],
        var_name = 'island', 
        value_name = 'type_of_island'
    )
)
```

```python
clear_df_penguins.head()
```

```python
clear_df_penguins.pivot_table(
        values = 'body_mass_g',
        index = 'species',
        columns = 'sex',
        aggfunc = 'count'
)
```

Гипотеза: количество самцов и самок в каждом виде примерно равны.
Фактически подтвердили гипотезу т.к. увидели что количество самцов и самок пингвинов по первым двум видам равно в выборке. По виду Gentoo больше женских особей но в пределах погрешности. И еще я дропнул пропуски в начале что могло повлиять на выборку.

```python
(
    clear_df_penguins
    .groupby(['sex', 'island'])
    .agg({'body_mass_g': ['mean', 'min', 'max', 'count']})
)
```

### Гипотеза1: самцы крупнее самок; Гипотеза 2: разнициы в размере пингвинов между островами нет
### Гипотеза 1 верна причем независимо от острова; Гипотеза 2 - неверна. Самцы и самки на острове Biscoe самые крупные
