### Взял Мцыри Лермонтова

```python
url = "https://lib.ru/LITRA/LERMONTOW/mcyri.txt_with-big-pictures.html"
response = requests.get(url)
soup = bs4.BeautifulSoup(response.text, 'html')
text = soup.body.get_text(' ', strip=True)
```

```python
text
```

```python
type(text)
```

```python
cl_text = text.lower().translate(str.maketrans('', '', string.punctuation)).translate(str.maketrans('', '', string.digits))
```

```python
cl_text
```

```python
with open("input.txt", "w", encoding = 'utf-8') as file:
    file.write(cl_text)
```

```python
data = db.read_text('input.txt', encoding = 'utf-8')
```

```python
data.compute()
```

```python
data.str.split().compute()
```

```python
type(data)
```

```python
words = data.str.split().flatten()
```

```python
words.compute()
```

```python
word_count = words.frequencies()
counts = pd.DataFrame(word_count, columns=['Word', 'Frequency']).sort_values(by="Frequency", ascending = False)
```

```python
counts.head(10)
```
